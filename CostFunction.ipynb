```python
import dataset
import numpy as np
import matplotlib.pyplot as plt
```

```python
# get data
datax, datay = dataset.getBeans(100)
print(datax[0:5], '...')
print(datay[0:5], '...')
```

```python
# graph properties
plt.title('Size-toxicity Graph', fontsize=12)
plt.xlabel('Bean Size')
plt.ylabel('Toxicity')
# add data points
plt.scatter(datax, datay)
plt.show()
```

```python
# cost function
ws = np.arange(0, 3, 0.1)
costs = []
for i in ws:
    # prediction
    output = i * datax
    # cost
    costs += [np.sum((datay - output)) ** 2 / 100]
# graph properties
plt.title('Cost Function', fontsize=12)
plt.xlabel('Weight')
plt.ylabel('Cost')
# graph function
plt.plot(ws, costs)

# lowest point
min_w = np.sum(datay * datax) / np.sum(datax ** 2)
print('Min weight:', min_w)

# display
plt.show()
```

```python
# graph properties
plt.title('Size-toxicity Graph', fontsize=12)
plt.xlabel('Bean Size')
plt.ylabel('Toxicity')
# add data points
plt.scatter(datax, datay)

outputs = min_w * datax
# graph function
plt.plot(datax, outputs)
plt.show()
```
