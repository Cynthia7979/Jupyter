```python
import numpy
import dataset
from matplotlib import pyplot as plt
```

```python
# get data
datax, datay = dataset.getBeans(100)
print(datax[0:5], '...')
print(datay[0:5], '...')
```

```python
# graph properties
plt.title('Size-toxicity Graph', fontsize=12)
plt.xlabel('Bean Size')
plt.ylabel('Toxicity')
# add data points
plt.scatter(datax, datay)

# alpha - learning rate
alpha = 0.01
# y = wx
w = 0.5
for i in range(100):
    for j in range(100):
        output = w * datax[j]    # predicted
        # calculate the differences
        differences = datay[j] - output
        # update the weight
        w = w + alpha * differences * datax[j]

# prediction
outputs = w * datax
# graph function
plt.plot(datax, outputs)
```
