```python
def draw_data(data,a,b):
    for i in range(data.data.shape[0]):
        plt.plot(data.data[i,a],data.data[i,b],'.',color=color[data.target[i]])
    plt.title('The shape of iris',size=20)
    plt.xlabel(data.feature_names[a],size=20) #title of x axis
    plt.ylabel(data.feature_names[b],size=20)
    plt.show()
                                                               
draw_data(data,0,3)
```

K-means starts with k centroids, associating the samples to the nearest centroid, then calculate the mean of each group to make k new centroids, which are then used in the next recursion of the steps above.

```python
model=KMeans(n_clusters=3) #3 clusters
model.fit(data.data) #feed data to the model
print((model.labels_ == data.target).sum()) #using == and .sum() on arrays,the first returning a boolean array 
print(model.labels_)
```

```python
plt.figure(figsize=(40,20))
color=['green','red','blue','yellow']

plt.subplot(1,2,1)
for i in range(150):
    plt.plot(data.data[i,0],data.data[i,1],'.',color=color[data.target[i]],markersize=20) #the data point for the sepal of i-th iris
plt.title('correct answer',size=20)
plt.xlabel('length',size=20) 
plt.ylabel('width',size=20)

plt.subplot(1,2,2)
for i in range(150):
    plt.plot(data.data[i,0],data.data[i,1],'.',color=color[model.labels_[i]],markersize=20) #the data point for the sepal of i-th iris
plt.title('answer by KMeans',size=20)
plt.xlabel('length',size=20) 
plt.ylabel('width',size=20)

plt.show()
```

`KNN` (K nearest neighbors) finds the k nearest samples of the test sample and returns the mostly occured label of the k neighbors.

```python
knnmodel=knn(12)#usually use the sqrt of sample size as the k
```

```python
test_index=np.random.randint(0,149,10)

test_set=[]
test_label=[]
train_set=[]
train_label=[]

for i in range(150):
    if i in test_index:
        test_label.append(data.target[i])
        test_set.append(data.data[i])
    else:
        train_label.append(data.target[i])
        train_set.append(data.data[i])

knnmodel.fit(train_set,train_label)
```

```python
(knnmodel.predict(test_set)==test_label).sum() #use .predict(test_set) to predict
```
