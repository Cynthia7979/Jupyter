```python
lst=[1,2,3,4,5,6]
arr=np.array(lst)
arr=arr**2 #yes arrays in np can be calculated like values in them
plt.plot(lst,arr,'-')
plt.show()
```

```python
print(np.random.random(size=10)) #rand from 0 to 1
print(np.random.randint(0,10,size=10)) #random integer from 0 to 10
```

```python
line=np.linspace(0,10,num=100000) #linspace, NOT linespace, returns an array of almost continuous values between the bounds; num refers to the size of the array
arr=np.sin(line)
plt.plot(line,arr,'-')
plt.show()
```

```python
def circle(x,y,r):
    c=np.linspace(0,6.3)
    a=x+r*np.sin(c)
    b=y+r*np.cos(c)
    plt.plot(a,b,'-')

plt.figure(figsize=[5,5]) #make x and y axis of the same length
for i in range(12):
    angle=i*np.pi/6
    circle(np.sin(angle),np.cos(angle),0.5)

plt.show()
```
