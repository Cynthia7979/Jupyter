# 目标检测
* 包括分类、检测和segmentation（语义分割）
* 传统目标检测（手工设计）
    * 区域选择
    * 特征提取
        * 常用HOG, SIFT等
    * 分类器
        * SVM（向量机）和自适应增强（Adaboot）等等
    * 优点；特定场景，较小类别能够有效检测
    * 缺点：时间复杂度过高，会产生冗余窗口
* 深度神经网络
    * 思路：将回归问题转为分类问题
    * 将图片分解成小框，将不同框的分类合在一起成为检测框
    * 历史：
        * 2012年跟随GPU发展
        * 2016年，YOLO开展单阶段（One-stage）检测器
            * 算法优点：快速，应用方便
            * 一次识别对象类型**和**区域
            * 双阶段：先生成区域，再识别对象
    * AlexNet
        * 在目标识别挑战上得到第一名
        * 引出了之后许多神经网络的发展
    * R-CNN
        * 目标检测鼻祖级论文
        * 由RBG团队提出
        * 开启了之后的神经网络发展
        * 结构：
            * 输入图片
            * 提取约2k张候选框
            * 区域变换（直接变换到需要的大小）
            * 输入CNN网络
            * 分类候选框（线性二分类，需要单独训练）
        * 解决的问题
            1. 如何利用神经网络进行目标定位？
                * 借用滑动窗口思想，提取独立候选区域
                * 对每个区域用CNN抽取固定长度的特征向量
                * 每个区域用SVM进行分类
    * Fast RCNN
        * 解决了训练/测试速度慢，训练所需空间大的问题
        * 准确率 ~66%-67%
        * 两个输入：图像和region proposal
        * 提出了ROI pooling的概念：在不规则区域内，提取每个小区域的最大值，最后输出需要的dimension
        * 将bonding box的回归部分整合进了神经网络中
        * 缺点：选择性搜索算法消耗很多时间
    * Faster RCNN
        * 分为四个部分
            1. 特征提取
                * 卷积层，和RPN共享参数
            2. RPN
                * Region Proposal Network
                * 引入多尺度Anchor
                * 分类背景/前景，更快速检测目标
            3. ROI感兴趣区域
            4. 分类和回归
        * 工作步骤
            * 在特征图上滑动窗口
            * 建立神经网络用于物体分类，框位置的回归
            * 滑动窗口的位置提供物体的位置信息
            * 框的回归提供了更精确的位置
    * R-FCN
        * 2016年发表
        * 基于RefNet-101
        * 精度高，速度快
    * Mask RCNN
        * 提供第三个分支，输出目标掩码，提供更加精确的语义分割
    * YOLO
        * 最后一层使用线性激活
        * 其他层采用Leaky RelUU
        * 划分为7*7的栅格
        * 包括v1, v2, v3，后面还有更多非官方版本
        * v2使用更多bounding box，提升位置准确性
    * SSD
        * 在YOLOv1后提出，用于提升精度（Yv1精度低于Faster RCNN）
        * 在图片上下添加padding
        * 分割成不同尺寸的网格
        * 物体的位置尺寸以能检测出来物体的网格决定
